```python
import sys
from pathlib import Path

ROOT_DIR = Path.cwd().parent

sys.path.insert(0, str(ROOT_DIR / "src"))

print(ROOT_DIR)
print((ROOT_DIR / "src").exists())
```

```python
SLANG_PATH = str(ROOT_DIR / "data" / "resources" / "slang_indo.csv")

print(SLANG_PATH)
```

```python
import pandas as pd

from preprocessing.pipeline import TextPreprocessor

pd.set_option('display.max_colwidth', 300)

print("Libraries loaded successfully.")
```

```python
# LOAD RAW DATA

TRAIN_PATH = "../data/raw/train.csv"
TEST_PATH = "../data/raw/test.csv"

train_df = pd.read_csv(TRAIN_PATH)
test_df = pd.read_csv(TEST_PATH)

print("Train Shape :", train_df.shape)
print("Test Shape  :", test_df.shape)
```

```python
# V1 - TFIDF - BASELINE
preprocessor_v1 = TextPreprocessor(
    use_slang_normalization = True,
    use_stemming            = True,
    use_stopword_removal    = True,
    remove_emoji            = True,
    slang_path = str(ROOT_DIR / "data" / "resources" / "slang_indo.csv")
)

# V2 - TRANSFORMER
preprocessor_v2 = TextPreprocessor(
    use_slang_normalization = True,
    use_stemming            = False,
    use_stopword_removal    = False,
    remove_emoji            = True,
    slang_path = str(ROOT_DIR / "data" / "resources" / "slang_indo.csv")
)
```

```python
# APPLY PREPROCESSING V1
train_df['clean_text_v1'] = train_df['full_text'].apply(
    preprocessor_v1.preprocess
)

test_df['clean_text_v1'] = test_df['full_text'].apply(
    preprocessor_v1.preprocess
)

print("V1 preprocessing completed.")
```

```python
# APPLY PREPROCESSING V2

train_df['clean_text_v2'] = train_df['full_text'].apply(
    preprocessor_v2.preprocess
)

test_df['clean_text_v2'] = test_df['full_text'].apply(
    preprocessor_v2.preprocess
)

print("V2 preprocessing completed.")
```

```python
# BEFORE VS AFTER

comparison_df = pd.DataFrame({
    
    'original_text': train_df['full_text'].head(10),
    
    'clean_text_v1': train_df['clean_text_v1'].head(10),
    
    'clean_text_v2': train_df['clean_text_v2'].head(10)
})

display(comparison_df)
```

```python
# SANITY CHECK

print("=== EMPTY TEXT CHECK ===")

print(
    "V1 Empty Text:",
    (train_df['clean_text_v1'].str.len() == 0).sum()
)

print(
    "V2 Empty Text:",
    (train_df['clean_text_v2'].str.len() == 0).sum()
)
```

```python
# SAVE INTERIM DATA

# SAVE V1

train_v1 = train_df[['clean_text_v1', 'label']].rename(
    columns={
        'clean_text_v1': 'full_text'
    }
)

test_v1 = test_df[['id', 'clean_text_v1']].rename(
    columns={
        'clean_text_v1': 'full_text'
    }
)


# SAVE V2

train_v2 = train_df[['clean_text_v2', 'label']].rename(
    columns={
        'clean_text_v2': 'full_text'
    }
)

test_v2 = test_df[['id', 'clean_text_v2']].rename(
    columns={
        'clean_text_v2': 'full_text'
    }
)


# EXPORT CSV

train_v1.to_csv(
    "../data/interim/train_clean_v1.csv",
    index=False
)

test_v1.to_csv(
    "../data/interim/test_clean_v1.csv",
    index=False
)

train_v2.to_csv(
    "../data/interim/train_clean_v2.csv",
    index=False
)

test_v2.to_csv(
    "../data/interim/test_clean_v2.csv",
    index=False
)

print("All interim datasets saved successfully.")
```
